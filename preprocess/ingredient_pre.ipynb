```python
import re
```

```python
file = open('../data/all_ingredients.txt','r')
data = file.read().split('\n')
file.close()

# data
```

```python
num_on_start_and_in_parentheses = r"\(.*?\)|^\d+[.)]\s"
```

```python
def clean_ingredients_text(ingredients):
  num_on_start_pattern = re.compile(num_on_start_and_in_parentheses)

  clean_ingredients = [num_on_start_pattern.sub('',ingredient) for ingredient in ingredients]
  return list(set(clean_ingredients))
```

```python
for ingredient in data:
  pattern = re.compile(r"\(.*?\)")
  if pattern.search(ingredient):
    print(ingredient,pattern.findall(ingredient))
```

```python
def classify_ingredients_array_by_key_word(ingredients, keyword):
  ingredients1 = []
  ingredients2 = []

  pattern = re.compile(keyword)

  for ingredient in ingredients: 
    if pattern.match(ingredient) or  pattern.search(ingredient):
      ingredients1.append(ingredient)
    else:
      ingredients2.append(ingredient)

  return ingredients1, ingredients2
```

```python
ingredients_with_egg_keyword, rest_data = classify_ingredients_array_by_key_word(data,'р╣Др╕Вр╣И')
```
