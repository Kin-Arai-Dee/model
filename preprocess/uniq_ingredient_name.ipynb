```python
import pandas as pd

df = pd.read_csv('../data/uniq_title_data.csv')
```

```python
from ast import literal_eval

df = df[['title','ingredients']]
df['ingredients'] = df["ingredients"].apply(literal_eval)
df
```

```python
all_ingredients = df["ingredients"].apply(lambda ingredients: [ingredient["ingredientName"] for ingredient in ingredients]).tolist()
```

```python
all_ingredients = list(set(sum(all_ingredients,[])))
```

```python
all_ingredients
```

```python
with open('../data/all_ingredients.txt', 'w') as f:
  f.write('\n'.join(all_ingredients))
  f.close()
```
