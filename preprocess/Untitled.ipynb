```python
df = pd.read_csv('../data/non.csv').sort_values(by=['frequency'], ascending=False).reset_index()
list_df = []
for i in range(len(df)):
    list_df.append([ df.loc[i]['ingredients'] , df.loc[i]['frequency'] ])
```

```python
all_facility_unique = [[i[0],i[1],0] for i in list_df if i[1] >= 1]
```

```python
def search_fsc(all_facility_unique, ch_regex):
    list_exam = []
    for i in range (len(all_facility_unique)) : 
        if re.search(ch_regex, all_facility_unique[i][0]):
            if all_facility_unique[i][1] > 1:
                print(all_facility_unique[i][0],all_facility_unique[i][1])
            all_facility_unique[i][2] = 1
            if all_facility_unique[i][1] > 1:
                list_exam.append(all_facility_unique[i][0])
    example = ', '.join(list_exam)
#     print(example)
    return all_facility_unique,example

def make_table(name,ch_regex,example):
    global df_name
    global df_ch_regex
    global df_example
    df_name.append(name)
    df_ch_regex.append(ch_regex)
    df_example.append(example)

df_name = []
df_ch_regex = []
df_example = []
```
