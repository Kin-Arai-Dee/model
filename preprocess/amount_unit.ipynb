```python
import re
def remove_emojis(data):
    emoji = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoji, '', data)
```

```python
def clean_text(text):
    cleaned_text = re.sub(r'[^\w\s\-\.\,\(\)\{\}\[\]\/\+\u0E00-\u0E7F]+|_', ' ', text)
    cleaned_text = re.sub(space_regex, ' ', cleaned_text)
    return cleaned_text
```

```python
def remove_text_to_unit(input_string):
    cleaned_text = clean_text(input_string)
    return cleaned_text
```

```python
def remove_leading_zeros(s: str):
    parts = s.split('.')
    if len(parts) == 2:
        left, right = parts
        if left == '0':
            left = ''
        return f"{left}.{right}"
    else:
        return s.lstrip('0') or '0'
```

```python
def replace_amount_word(s: str):
    s = s.replace('ครึ่ง','.5')
    s = s.replace('ส่วน','/')
    s = s.replace('ใน','/')
    s = s.replace('X','*')
    s = s.replace('x','*')
    s = s.replace('ถึง','-')
    s = s.replace('0้อนโต๊ะ','ช้อนโต๊ะ')
    s = s.replace('ประมาณ','')
    s = s.replace('12. ข้าวสวย 150-200 กรัม','150-200 กรัม')
    s = s.replace('2-3 ต้น / 2-3 ช้อนโต๊ะ','2-3 ช้อนโต๊ะ')
    s = split_string_by_word(s,r'หรือ|\(|\)')
    return s
```

```python
def replace_name_word(s: str,p):
    if s  == 'กระเทียมกรีบ':
        return (s,'1 กรีบ')
    if s.startswith('Lobster 0.7-1kg'):
        return ('Lobster','0.7-1kg')
    if s  == 'ปูเนื้อ (ตัวนี้หนักประมาณ 7 ขีด) 1 ตัว':
        return ('ปูเนื้อ','ขึด')
    if s == 'เนื้อปลาอินทรีย์หนักประมาณ 2 ขีด':
        return ('เนื้อปลาอินทรีย์','2 ขึด')
    if s == 'คอหมูย่าง ประมาณ 4 ขีด':
        return ('คอหมูย่าง','4 ขึด')

    s = re.sub(r"^\d+[.)|\s]*",'',s)
    s = re.sub(r'\(([^)]+)\)','',s)
    s = s.replace('2-3 ต้น / 2-3 ช้อนโต๊ะ','2-3 ช้อนโต๊ะ')
    s = s.replace('2-3 ต้น / 2-3 ช้อนโต๊ะ','2-3 ช้อนโต๊ะ')
    s = split_name_by_word(s,p)
    return s, None
```

```python
def recur_change_to_number(number_list):
  if not number_list:
    return 0
  if '-' in number_list:
    dash_index = number_list.index('-')
    return (recur_change_to_number(number_list[:dash_index]) + recur_change_to_number(number_list[dash_index+1:]))/2
  number_list2 = []
  for num in number_list:
    try:
      number_list2.append(remove_leading_zeros(str(eval(num))))
    except:
      number_list2.append(num.strip())
  try:
    return eval(''.join(number_list2))
  except:
    # print(number_list2)


    if '+' == number_list2[-1]:
      recur_change_to_number(number_list2[:-1])
    else:
      # print('+'.join(number_list2))
      return eval('+'.join(number_list2))

# recur_change_to_number(['1','1/2'])
```

```python
def amount_value(number_matchs):
  original = number_matchs
  for i, match_str in enumerate(number_matchs):
    new_value = ''
    for match in match_str:
      if match in fraction_mapper:
        new_value += fraction_mapper[match]
      else:
        new_value += match
    number_matchs[i] = new_value.strip()

  value =  recur_change_to_number(number_matchs)
  printf = False
  if len(original) > 1 and ('+' not in original and '-' not in original) :
    # print(original, value)
    printf = True
  return value, printf

# amount_value(['1','+','¼','+','1+⅞','-','1⅞','+','⅞'])
# amount_value(['.0','.5'])
```

```python
count = 0
c = 0

p_list = []
def split_size_generater(ingredient):
  global count 
  global p_list
   
  sizes = []

  p_it = False

  for x in ingredient:

    name, amount = x
    name, new_amount = replace_name_word(name,amount)
    # fraction_regex = re.findall(r'[\u00BC-\u00BE\u2150-\u215E]+',x)
    if new_amount:
      amount = new_amount
    amount = amount.replace('\u200b','')
    amount = replace_amount_word(amount)

    
    number_matchs = re.findall(r'((?:[\d\u00BC-\u00BE\u2150-\u215E]+(?:\.\d+)?(?:\s*[\/]\s*\d+)?[\u00BC-\u00BE\u2150-\u215E]*)|(?:[\+\-\*])|(?:\.\d+))',amount)
    # if len(matchs) > 1:
    units = re.sub(r'((?:[\d\u00BC-\u00BE\u2150-\u215E]+(?:\.\d+)?(?:\s*[\/]\s*\d+)?[\u00BC-\u00BE\u2150-\u215E]*)|(?:[\+\-\*])|(?:\.\d+))','',amount).strip()

    value, p = amount_value(number_matchs)

    if p: 
      p_it = p
    # if(len(number_matchs)>2):
    #   print(count, ':', value, "|",x)
    sizes.append((name,value,units))

  count += 1
  if(p_it):
    p_list.append(count)
  # print(sizes)

  return sizes
```
