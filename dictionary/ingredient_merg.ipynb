```python
def create_dict_with_matched_values(lst, pattern_key_pairs):
    matched_values = {}
    for pattern, key_name in pattern_key_pairs:
        matched_elements = []
        regex = re.compile(pattern)
        for element in lst:
            if regex.match(element):
                matched_elements.append(element)
        for element in matched_elements:
            lst.remove(element)
        if matched_elements:
            matched_values[key_name] = matched_elements
    return matched_values
```
