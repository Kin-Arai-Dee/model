```python
import pandas as pd


df = pd.read_csv('../data/data_with_title.csv')

# df
```

```python
translate_list = df[['title','short_tile']].to_dict(orient='index').values()
```

```python
d = {}

for word in translate_list:
  d[word['title'].strip()] = word['short_tile']
```

```python
d
```

```python
import json

with open('../dictionary/title_dict.json','w',encoding='utf8')as file:
  json.dump(d,file,ensure_ascii=False)
```
