```python
import re

def is_any_match_regex(x,pattern,name):
  if(any([re.search(pattern, s) for s in x])):
    return name
  else: return ''
  
```

```python
df['uniq_unit'] = df['unit']
```

```python
def uniq_unit_by_regex(x: list[str],pattern,name):
  is_match = False
  new_x = []
  for s in x:
    if re.search(pattern,s.lower()):
      is_match = True
      continue
    else:
      new_x.append(s)
  if is_match:
    new_x.append(name)
  return new_x
```

```python
for key,value in unit_dict.items():
  df[key] = df['unit'].apply(lambda x: is_any_match_regex(x,value,key))
  df['uniq_unit'] = df['uniq_unit'].apply(lambda x: uniq_unit_by_regex(x,value,key))
```
